```python
import pandas as pd
import numpy as np
```

```python
# Importando o dataset

df = pd.read_csv('saude_do_sono_estilo_vida.csv')

df.head(10)
```

```python
# 1. Ao visualizar a base você percebeu que seria melhor alterar o nome de algumas colunas. Mude o ‘ID’ para ‘Identificador’, corrija o nome da
# coluna que indica a pressão sanguínea, mude a coluna ‘Ocupação’ para ‘Profissão’, a coluna ‘Categoria BMI’ está em parte em inglês, substitua para ‘Categoria IMC’.

novo_nome = {'ID': 'Identificador',
             'Pressão sanguíneaaaa': 'Pressão Sanguínea',
             'Ocupação': 'Profissão',
             'Categoria BMI': 'Categoria IMC'
            }

df.rename(columns = novo_nome, inplace = True)

df.head(10)
```

```python
# 2. Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões? [‘mean’, np.median, pd.Series.mod]

df_sono_media = df.groupby('Profissão')['Duração do sono'].mean() # Média
df_sono_media
```

```python
df_sono_mediana = df.groupby('Profissão')['Duração do sono'].median() # Mediana
df_sono_mediana
```

```python
df_sono_moda = df.groupby('Profissão')['Duração do sono'].agg(pd.Series.mode) # Moda
df_sono_moda
```

```python
# 3. Das pessoas que atuam com engenharia de software qual a porcentagem de obesos?

df_profissao = df[df['Profissão'] == 'Eng. de Software']

achando_num = df_profissao[df_profissao['Categoria IMC'] == 'Obesidade'].shape[0]

resultado = (achando_num / len(df_profissao)) *100

print("A porcentagem de pessoas obesas em engenharia de software é:", resultado)

```

```python
# 4. De acordo com os dados, advogar ou ser representante de vendas faz você dormir menos? (Use o método ‘isin’, considere a média)

df_menos_sono = df[df['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])]

df_menos_sono = df_menos_sono['Duração do sono'].mean()
```

```python
media_sono_geral = sum(df_sono_media) / len(df_sono_media)

if df_menos_sono < media_sono_geral:
  print('Advogar ou ser representante de vendas faz você dormir menos que a média geral')
else:
  print('Advogar ou ser representante de vendas faz você dormir mais que a média geral')
```

```python
# 5. Entre quem fez enfermagem e quem fez medicina, quem tem menos horas de sono? (Use o método ‘isin’, considere a média)

df_enfermeiro = df[df['Profissão'].isin(['Enfermeiro(a)'])]['Duração do sono'].mean()
df_medico = df[df['Profissão'].isin(['Médico(a)'])]['Duração do sono'].mean()

if df_medico > df_enfermeiro:
  print('Quem fez enfermagem tem menos horas de sono.')
elif df_enfermeiro > df_medico:
  print('Quem fez medicina tem menos horas de sono.')
```

```python
# 6. Faça um subconjunto com as colunas Identificador, Gênero, Idade, Pressão sanguínea e Frequência cardíaca.

subconjunto = df[['Identificador', 'Gênero', 'Idade', 'Pressão Sanguínea', 'Frequência cardíaca']]
subconjunto.head(10)
```

```python
# 7. Descubra qual a profissão menos frequente no conjunto. (Use value_counts)

menos_frequente =  df['Profissão'].value_counts()
profissao_menos_frequente = menos_frequente.idxmin()
print('A profissão menos frequente no conjunto é:', profissao_menos_frequente)
```

```python
# 8. Quem tem maior pressão sanguínea média, homens ou mulheres? (Considere a média)

df[['Sistólico', 'Diastólico']] = df['Pressão Sanguínea'].str.split('/', expand=True).astype(int)

media_homem = df[df['Gênero'] == 'Homem']['Sistólico'].mean(), df[df['Gênero'] == 'Homem']['Diastólico'].mean()
media_mulher = df[df['Gênero'] == 'Mulher']['Sistólico'].mean(), df[df['Gênero'] == 'Mulher']['Diastólico'].mean()

if media_homem > media_mulher:
  print('Homens tem uma pressão sanguínea maior.')
elif media_homem < media_mulher:
  print('Mulheres tem uma pressão sanguínea maior.')


```

```python
# 9. É predominante entre os participantes dormir 8 horas por dia (considere usar Moda como medida)?

df_dormir = df['Duração do sono'].mode()[0]
print('Não são predominantes os partipantes que dormem 8 horas por dia, mas sim os que dormem', df_dormir, 'por dia.')
```

```python
# 10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

passos_acima = df[df['Frequência cardíaca'] > 70]['Passos diários'].mean()
passos_abaixo = df[df['Frequência cardíaca'] <= 70]['Passos diários'].mean()

if passos_acima > passos_abaixo:
  print('Pessoas com frequência cardíacas acima de 70 dão mais passos.')
elif passos_acima < passos_abaixo:
  print('Pessoas com frequência cardíacas menor ou igual a 70 dão mais passos.')
```
